# Manual way of adding font

Download the font file and attach to the fontproperties with the path

```python
# Library & Module
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
# Path
font_prop = FontProperties(fname='path_to_font_file')
```

```python
# Dummy Plot
nepali_font = FontProperties(family='Nepali Font')
plt.text(0.5, 0.5, 'नेपाली', fontproperties=font_prop)
plt.show()
```
